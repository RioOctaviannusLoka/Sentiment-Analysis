```python
from google_play_scraper import app, reviews_all, Sort
import pandas as pd
```

# Data Scraping

```python
scrap_review = reviews_all(
    'com.mobile.legends',
    lang='id',
    country='id',
    sort=Sort.MOST_RELEVANT,
    count=80000
)
```

# Check Dataset

```python
df = pd.DataFrame(scrap_review)
df.head()
```

```python
df.shape
```

# Store Dataset

```python
df.to_csv('dataset.csv', index=False)
```
